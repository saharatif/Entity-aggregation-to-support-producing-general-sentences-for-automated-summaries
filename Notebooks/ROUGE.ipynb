```python
from rouge import Rouge
```

NYT ROUGE

```python
refs=[]
hyps=[]
ref_file= open('79_nyt_orig_summ_new.txt','r')
for line in ref_file:
    stripped_line = line.strip()
    refs.append(stripped_line)
    
ref_file= open('79_t5_single_sentsum_new.txt','r')
for line in ref_file:
    stripped_line = line.strip()
    hyps.append(stripped_line)    
```

```python
rouge = Rouge()
scores = rouge.get_scores(hyps, refs, avg=True)
```

```python
print(scores)
```

CNN ROUGE

```python
refs=[]
hyps=[]
ref_file= open("94-cnn_orginal_summaries_single_sent.txt",'r')
for line in ref_file:
    stripped_line = line.strip()
    refs.append(stripped_line)
    
ref_file= open('94-cnn_t5_single_sentsum.txt','r')
for line in ref_file:
    stripped_line = line.strip()
    hyps.append(stripped_line) 
```

```python
rouge = Rouge()
scores = rouge.get_scores(hyps, refs, avg=True)
scores
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = [12, 8]
plotdata = pd.DataFrame({
    "Rouge-1":[0.572, 0.739, 0.633],
    "Rouge-2":[0.433, 0.595, 0.490],
    "Rouge-l":[0.559, 0.721, 0.618]
    }, 
    index=["Recall", "Precision", "F-1 score"]
)


font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 16,
        }

list_color=['lightsteelblue','lightskyblue','royalblue']
plotdata.plot(kind="bar",color=list_color)
plt.title("Rouge score for NYT Annotated Corpus",fontdict=font)
plt.xticks(rotation = 45)
plt.xlabel("")
plt.ylabel("Rouge values",fontdict=font)
plt.savefig('NYT_rouge.png')
```
