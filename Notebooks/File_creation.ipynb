Extracting from the original file nyt50

```python
import jsonlines
import pandas as pd
import re
article_list=[]
summary_list=[]
define_words = 'and other '
# file=open("NYT_summaries.txt_v1","w")
with jsonlines.open("nyt50.ext.outputs", 'r') as reader:
    for example in reader:
        system_summary = example['system summary']
        article = example['src']
        for match in re.findall(r"(?m)^.*?\band other\b.*?$" ,example['system summary']):
#             print('\n\n')
            summary=match.replace('[CLS]',"").replace('<q>',"")
            article=example["src"].replace('[CLS]',"").replace('<q>',"").replace('[SEP]',"")
            article_list.append(article)
            summary_list.append(summary)
#             print(summary)
#             print('\n\n')
#             print(article)

df_and_other = pd.DataFrame(list(zip(article_list,summary_list)),
               columns =['Article', 'Summary'])
df_and_other
```

Make Hearst file from Summary file

```python
file_such_as=open("such_as_sentences_NYT_sum.txt","w")
import re
define_words = 'such as '
for i, line in enumerate(open('NYT_summaries.txt','r')):
    for match in re.findall(r"([^.]*?%s[^.]*\.)" % define_words,line):
        file_such_as.write(str(match))
        file_such_as.write('\n')
file_such_as.close() 
```

```python
file_and_other=open("and_other_sentences_NYT_sum.txt","w")
import re
define_words = 'and other '
for i, line in enumerate(open('NYT_summaries.txt','r')):
    for match in re.findall(r"([^.]*?%s[^.]*\.)" % define_words,line):
        file_and_other.write(str(match))
        file_and_other.write('\n')
file_and_other.close() 
```
