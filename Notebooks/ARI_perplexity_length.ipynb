```python
def ari(text):
    ari_text=textstat.automated_readability_index(text)
    
    return ari_text
    
```

```python
def perplexity(text):
    
    model = FlairEmbeddings('news-forward').lm
    input = torch.tensor([model.dictionary.get_idx_for_item(char) for char in text[:-1]]).unsqueeze(1)

    # push list of character IDs through model
    hidden = model.init_hidden(1)
    prediction, _, hidden = model.forward(input, hidden)

    # the target is always the next character
    targets = torch.tensor([model.dictionary.get_idx_for_item(char) for char in text[1:]])

    # use cross entropy loss to compare output of forward pass with targets
    cross_entroy_loss = torch.nn.CrossEntropyLoss()
    loss = cross_entroy_loss(prediction.view(-1, len(model.dictionary)), targets).item()

    # exponentiate cross-entropy loss to calculate perplexity
    perplexity = math.exp(loss)

    return perplexity
```

PERPLEXITY OF T5

```python
from transformers import AutoModelWithLMHead, AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("mrm8488/t5-base-finetuned-summarize-news")
model = AutoModelWithLMHead.from_pretrained("mrm8488/t5-base-finetuned-summarize-news")
```

```python
import numpy as np
def perplex(text, max_length=150):
    input_ids = tokenizer.encode(text, return_tensors="pt", add_special_tokens=True)
    labels = tokenizer(text, return_tensors="pt").input_ids

    generated_ids = model.generate(input_ids=input_ids, num_beams=2, max_length=max_length,  repetition_penalty=2.5, length_penalty=1.0, early_stopping=True)

    preds = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=True) for g in generated_ids]
    loss = model(input_ids=input_ids, labels=labels).loss
    loss=torch.exp(loss)
#     loss=loss.detach().numpy()[0]

    return loss.item()
```

```python
text="""gov . eliot spitzer will preview his first executive budget in a speech on friday afternoon , laying out his plan to rein in medicaid and other health care spending , setting the stage for a battle with politically powerful health care providers and unions ."""


print(perplex(text))
```
