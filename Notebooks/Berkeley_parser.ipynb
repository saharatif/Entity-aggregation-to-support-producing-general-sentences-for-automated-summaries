```python
import benepar, spacy
nlp = spacy.load('en_core_web_sm')
if spacy.__version__.startswith('2'):
    nlp.add_pipe(benepar.BeneparComponent("benepar_en3"))
else:
    nlp.add_pipe("benepar", config={"model": "benepar_en3"})
```

```python
#Extract sentences with pattern
file_parse=open("parsed_sentences.txt","a+")
import re
define_words = 'and other '
for i, line in enumerate(open('CNN_articles.txt','r')):
    for match in re.findall(r"([^.]*?%s[^.]*\.)" % define_words,line):
        doc=nlp(match)
        sent = list(doc.sents)[0]
        print(sent)
        print(sent._.parse_string)
        print("\n")
        file_parse.write(str(sent))
        file_parse.write('\n')
        file_parse.write(sent._.parse_string)
        file_parse.write('\n')
        file_parse.write('\n')
file_parse.close()
```
